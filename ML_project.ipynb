```python
loan_df["Gender"] = loan_df["Gender"].fillna(loan_df["Gender"].iloc[0])
```

```python
loan_df.head()
```

```python
loan_df["Dependents"] = loan_df["Dependents"].replace('3+','3',regex = False)
loan_df.head()
```

```python
# Fill NA values with the mean
loan_df["Dependents"] = loan_df["Dependents"].fillna(loan_df["Dependents"].iloc[0])
```

```python
# Fill missing values in the "Self_Employed" column:This code replaces all NaN (missing) values in the "Self_Employed" column 
loan_df["Self_Employed"] = loan_df["Self_Employed"].fillna(loan_df["Self_Employed"].iloc[0])
```

```python
# Fill missing values in the "Credit_History" column:This code replaces all NaN (missing) values in the "Credit_History" 
loan_df["Credit_History"] = loan_df["Credit_History"].fillna(loan_df["Credit_History"].iloc[0])
```

```python
loan_df.head()
```

```python
# Fill missing values in the "Loan Amount" column:This code replaces all NaN (missing) values in the "Loan Amount" column 

loan_df["LoanAmount"] = loan_df["LoanAmount"].fillna(loan_df["LoanAmount"].mean())
```

```python
loan_df.head()
```

```python
# Fill missing values in the "Loan_Amount_Term" column:This code replaces all NaN (missing) values in the "Loan_Amount_Term" column 
loan_df["Loan_Amount_Term"] = loan_df["Loan_Amount_Term"].fillna(loan_df["Loan_Amount_Term"].iloc[0])
```

```python
loan_df.head()
```

```python
# Fill missing values in the "Loan_Status" column:This code replaces all NaN (missing) values in the "Loan_Status" column
loan_df["Loan_Status"] = loan_df["Loan_Status"].fillna(loan_df["Loan_Status"].iloc[0]).astype(int)
```

```python
loan_df.head()
```

```python

Income_group_df = loan_df.groupby("Gender")['ApplicantIncome'].agg(["max", "min", "mean"]).reset_index()
```

```python
Income_group_df
```

```python
from sklearn.preprocessing import LabelEncoder

# List of columns to encode
Loan_df_encode = ["Education", "Married", "Gender", "Self_Employed", "Property_Area"]

# Apply LabelEncoder to each column
for col in Loan_df_encode:
    loan_df[f"{col}Label"] = LabelEncoder().fit_transform(loan_df[col])
```

```python
loan_df.head()
```

```python
# Save CSV file 
loan_df.to_csv(r"C:\Users\tades\OneDrive\Desktop\ML project\LoanData.csv", index = False)
```

```python
x =  loan_df[["ApplicantIncome","CoapplicantIncome","LoanAmount","EducationLabel","MarriedLabel","GenderLabel"]]
y = loan_df["Loan_Status"]
```

```python
x
```

```python
y
```

```python
#Train and test data
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.2)
```

```python
x_train.shape
```

```python
x_test.shape
```

```python
# Creating the model
from sklearn.linear_model import LogisticRegression
reg = LogisticRegression()
reg.fit(x_train,y_train)
```

```python
reg.score(x_train, y_train)
```

```python
reg.score(x_test, y_test)
```

## The training and testing accuracies of the logistic regression model 82.59%  for training and 87.84% for testing) suggest that the model performs reasonably well in predicting loan eligibility.
