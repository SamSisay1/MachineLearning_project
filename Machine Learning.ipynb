```python
import numpy as np

# Data points
x = np.array([5, 7, 8, 7, 2, 17, 2, 9, 4, 11, 12, 9, 6])
y = np.array([99, 86, 87, 88, 111, 86, 103, 87, 94, 78, 77, 85, 86])

# Calculate means
mean_x = np.mean(x)
mean_y = np.mean(y)

# Calculate components of the slope formula
n = len(x)
sum_xy = np.sum(x * y)
sum_x = np.sum(x)
sum_y = np.sum(y)
sum_x2 = np.sum(x**2)

# Slope (m) calculation
slope = (n * sum_xy - sum_x * sum_y) / (n * sum_x2 - sum_x**2)

# Y-intercept (c) calculation
y_intercept = mean_y - slope * mean_x

slope, y_intercept
```
